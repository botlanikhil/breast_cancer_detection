```python
# Cell 1: Install packages and mount drive
!pip install tensorflow keras numpy opencv-python matplotlib scikit-learn flask

from IPython import get_ipython
from IPython.display import display
import os


# Cell 2: Import libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3, EfficientNetB0
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Concatenate, GlobalAveragePooling2D, Input
from tensorflow.keras.models import Model
from tensorflow.keras.utils import Sequence
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import cv2
from sklearn.model_selection import train_test_split

# Cell 3: List directories to debug path
project_path = "/content/drive/MyDrive/cancer project"
if os.path.exists(project_path):
    print("\nDirectories in cancer project:")
    for item in os.listdir(project_path):
        print(item)
else:
    print("cancer project does not exist")

# Cell 4: Verify dataset path
dataset_path = "/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT"
if os.path.exists(dataset_path):
    print(f"\ndataset path is correct and contains: {os.listdir(dataset_path)}")
else:
    print("dataset path does not exist, update the path")

# Cell 5: Image parameters
IMG_SIZE = 224
BATCH_SIZE = 16
EPOCHS = 10

# Cell 6: Image Augmentation - Fixed validation split issue
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.6, 1.4],
    fill_mode="nearest"
)

val_datagen = ImageDataGenerator(rescale=1.0 / 255.0)

# Cell 7: Load training & validation data - Fixed data loading
# Assuming you have separate train/val folders or create them
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    dataset_path,  # You might need separate validation folder
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

print(f"Number of classes: {train_generator.num_classes}")
print(f"Class indices: {train_generator.class_indices}")

# Cell 8: Create ensemble model with proper input handling
def create_ensemble_model(num_classes):
    # Single input layer
    input_layer = Input(shape=(224, 224, 3))
    
    # Load pretrained models
    inception_base = InceptionV3(weights='imagenet', include_top=False, input_tensor=input_layer)
    efficientnet_base = EfficientNetB0(weights='imagenet', include_top=False, input_tensor=input_layer)
    
    # Freeze pretrained layers initially
    inception_base.trainable = False
    efficientnet_base.trainable = False
    
    # Extract features
    inception_features = GlobalAveragePooling2D()(inception_base.output)
    efficientnet_features = GlobalAveragePooling2D()(efficientnet_base.output)
    
    # Merge features
    merged = Concatenate()([inception_features, efficientnet_features])
    merged = BatchNormalization()(merged)
    merged = Dense(512, activation='relu')(merged)
    merged = Dropout(0.3)(merged)
    merged = Dense(256, activation='relu')(merged)
    merged = Dropout(0.2)(merged)
    output = Dense(num_classes, activation='softmax')(merged)
    
    model = Model(inputs=input_layer, outputs=output)
    return model

# Create the model
model = create_ensemble_model(train_generator.num_classes)

# Cell 9: Compile model with correct loss function
model.compile(
    optimizer=Adam(learning_rate=0.0001), 
    loss='categorical_crossentropy',  # Fixed: was binary_crossentropy
    metrics=['accuracy']
)

model.summary()

# Cell 10: Train the model (initial training)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=[early_stopping]
)

# Cell 11: Fine-tuning - Unfreeze some layers
# Unfreeze last 20 layers of both models
for layer in model.layers:
    if hasattr(layer, 'layers'):  # Check if it's a nested model
        for sublayer in layer.layers[-20:]:
            sublayer.trainable = True

# Compile with lower learning rate
model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Cell 12: Fine-tuning training
fine_tune_epochs = 10
callbacks = [
    EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
]

history_fine = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=fine_tune_epochs,
    callbacks=callbacks
)

# Cell 13: Plot training results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
if 'history' in locals():
    plt.plot(history.history['accuracy'], label='Initial Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Initial Val Accuracy')
plt.plot(history_fine.history['accuracy'], label='Fine-tune Train Accuracy')
plt.plot(history_fine.history['val_accuracy'], label='Fine-tune Val Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
if 'history' in locals():
    plt.plot(history.history['loss'], label='Initial Train Loss')
    plt.plot(history.history['val_loss'], label='Initial Val Loss')
plt.plot(history_fine.history['loss'], label='Fine-tune Train Loss')
plt.plot(history_fine.history['val_loss'], label='Fine-tune Val Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()

# Cell 14: Create test set for evaluation
# Get all image paths and labels
image_paths = train_generator.filepaths
labels = train_generator.labels

# Convert categorical labels to one-hot if needed
from tensorflow.keras.utils import to_categorical
labels_categorical = to_categorical(labels, num_classes=train_generator.num_classes)

# Load and preprocess images
X = []
for image_path in image_paths:
    img = cv2.imread(image_path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    X.append(img)

X = np.array(X, dtype='float32') / 255.0

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, labels_categorical, test_size=0.2, random_state=42, stratify=labels
)

# Cell 15: Evaluate model
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# Cell 16: Save model
model.save('breast_cancer_ensemble_model.keras')
print("Model saved successfully!")

```

```python
# Cell 17: Grad-CAM visualization function
def get_gradcam(model, img_array, class_index=None):
    # Get the last convolutional layer by checking layer names and types
    last_conv_layer = None
    
    # First, try to find conv layers in the base models
    for layer in model.layers:
        if hasattr(layer, 'layers'):  # This is a nested model (InceptionV3 or EfficientNet)
            for sublayer in reversed(layer.layers):
                if 'conv' in sublayer.name.lower() and hasattr(sublayer, 'output'):
                    try:
                        output_shape = sublayer.output.shape
                        if len(output_shape) == 4:  # Check if it's a conv layer
                            last_conv_layer = sublayer
                            break
                    except:
                        continue
            if last_conv_layer:
                break
    
    # If no conv layer found, try a different approach
    if last_conv_layer is None:
        # Look for layers with 'conv' in name
        for layer in reversed(model.layers):
            if hasattr(layer, 'name') and 'conv' in layer.name.lower():
                last_conv_layer = layer
                break
        
        # If still not found, use a specific layer name that should exist
        if last_conv_layer is None:
            try:
                # Try common layer names from InceptionV3/EfficientNet
                possible_names = ['mixed10', 'mixed9', 'top_conv', 'block7a_expand_conv']
                for name in possible_names:
                    try:
                        last_conv_layer = model.get_layer(name)
                        break
                    except:
                        continue
            except:
                pass
    
    if last_conv_layer is None:
        print("No suitable convolutional layer found for Grad-CAM")
        print("Available layers:")
        for i, layer in enumerate(model.layers):
            print(f"{i}: {layer.name} - {type(layer).__name__}")
        return None
    
    print(f"Using layer: {last_conv_layer.name}")
    
    # Create gradient model
    try:
        grad_model = tf.keras.models.Model(
            model.inputs, [last_conv_layer.output, model.output]
        )
    except Exception as e:
        print(f"Error creating gradient model: {e}")
        return None
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if class_index is None:
            class_index = tf.argmax(predictions[0])
        loss = predictions[:, class_index]
    
    # Compute gradients
    grads = tape.gradient(loss, conv_outputs)
    if grads is None:
        print("Could not compute gradients")
        return None
        
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Generate heatmap
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0)
    
    # Normalize heatmap
    max_val = tf.reduce_max(heatmap)
    if max_val > 0:
        heatmap = heatmap / max_val
    
    return heatmap.numpy()

# Cell 18: Visualize Grad-CAM
def visualize_gradcam(image_path):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return
    
    # Load and preprocess image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    # Get prediction
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class]
    
    # Generate Grad-CAM
    heatmap = get_gradcam(model, img_array, predicted_class)
    
    if heatmap is not None:
        # Resize heatmap
        heatmap = cv2.resize(heatmap, (img.size[0], img.size[1]))
        heatmap = np.uint8(255 * heatmap)
        heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
        
        # Overlay on original image
        original_img = np.array(img)
        superimposed_img = cv2.addWeighted(original_img, 0.6, heatmap, 0.4, 0)
        
        # Plot results
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.imshow(original_img)
        plt.title("Original Image")
        plt.axis('off')
        
        plt.subplot(1, 3, 2)
        plt.imshow(heatmap)
        plt.title("Grad-CAM Heatmap")
        plt.axis('off')
        
        plt.subplot(1, 3, 3)
        plt.imshow(superimposed_img)
        plt.title(f"Overlay\nPredicted: Class {predicted_class}\nConfidence: {confidence:.2f}")
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()

# Cell 19: Alternative simpler Grad-CAM function
def simple_gradcam(model, img_array, layer_name=None):
    """
    Simplified Grad-CAM implementation that works better with complex models
    """
    # If no layer specified, try to find a suitable one
    if layer_name is None:
        # Try common layer names
        possible_layers = ['mixed10', 'mixed9', 'mixed8', 'top_conv', 'block7a_expand_conv', 'block6a_expand_conv']
        for name in possible_layers:
            try:
                test_layer = model.get_layer(name)
                layer_name = name
                print(f"Found layer: {layer_name}")
                break
            except:
                continue
    
    if layer_name is None:
        print("Could not find suitable layer. Available layers:")
        for layer in model.layers:
            print(f"- {layer.name}")
        return None
    
    # Create a model that maps the input image to the activations of the last conv layer
    grad_model = tf.keras.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )
    
    # Compute the gradient of the top predicted class for our input image
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    
    # Gradient of the output neuron (top predicted) with regard to the output feature map
    grads = tape.gradient(class_channel, last_conv_layer_output)
    
    # Mean intensity of the gradient over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Multiply each channel in the feature map array by "how important this channel is"
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    # Normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# Cell 20: Updated visualization function
def visualize_gradcam_simple(image_path, layer_name=None):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return
    
    # Load and preprocess image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    # Get prediction
    prediction = model.predict(img_array, verbose=0)
    predicted_class = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class]
    
    # Get class names if available
    class_names = list(train_generator.class_indices.keys()) if 'train_generator' in globals() else [f"Class {i}" for i in range(len(prediction[0]))]
    
    print(f"Predicted: {class_names[predicted_class]} (Confidence: {confidence:.3f})")
    
    # Generate Grad-CAM
    heatmap = simple_gradcam(model, img_array, layer_name)
    
    if heatmap is not None:
        # Resize heatmap to match original image size
        heatmap_resized = cv2.resize(heatmap, (img.size[0], img.size[1]))
        heatmap_colored = np.uint8(255 * heatmap_resized)
        heatmap_colored = cv2.applyColorMap(heatmap_colored, cv2.COLORMAP_JET)
        
        # Convert original image to array
        original_img = np.array(img)
        
        # Superimpose the heatmap on original image
        superimposed_img = heatmap_colored * 0.4 + original_img * 0.6
        superimposed_img = np.clip(superimposed_img, 0, 255).astype(np.uint8)
        
        # Display results
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.imshow(original_img)
        plt.title("Original Image")
        plt.axis('off')
        
        plt.subplot(1, 3, 2)
        plt.imshow(heatmap, cmap='jet')
        plt.title("Grad-CAM Heatmap")
        plt.axis('off')
        
        plt.subplot(1, 3, 3)
        plt.imshow(superimposed_img)
        plt.title(f"Overlay\n{class_names[predicted_class]}\nConf: {confidence:.3f}")
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
    else:
        print("Could not generate Grad-CAM visualization")

# Example usage - try both functions
sample_image_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/normal/normal (1).png'

print("Trying original Grad-CAM function:")
try:
    visualize_gradcam(sample_image_path)
except Exception as e:
    print(f"Error with original function: {e}")

print("\nTrying simplified Grad-CAM function:")
try:
    visualize_gradcam_simple(sample_image_path)
except Exception as e:
    print(f"Error with simplified function: {e}")

# If both fail, let's debug the model structure
print("\nModel structure debugging:")
print("Model layers:")
for i, layer in enumerate(model.layers):
    print(f"{i}: {layer.name} - {type(layer).__name__}")
    if hasattr(layer, 'layers') and layer.layers:
        print(f"   Sublayers: {len(layer.layers)} layers")
        # Show last few sublayers
        for j, sublayer in enumerate(layer.layers[-5:]):
            print(f"     {len(layer.layers)-5+j}: {sublayer.name} - {type(sublayer).__name__}")
```

```python
# Cell 17: Grad-CAM visualization function
def get_gradcam(model, img_array, class_index=None):
    # Get the last convolutional layer by checking layer names and types
    last_conv_layer = None
    
    # First, try to find conv layers in the base models
    for layer in model.layers:
        if hasattr(layer, 'layers'):  # This is a nested model (InceptionV3 or EfficientNet)
            for sublayer in reversed(layer.layers):
                if 'conv' in sublayer.name.lower() and hasattr(sublayer, 'output'):
                    try:
                        output_shape = sublayer.output.shape
                        if len(output_shape) == 4:  # Check if it's a conv layer
                            last_conv_layer = sublayer
                            break
                    except:
                        continue
            if last_conv_layer:
                break
    
    # If no conv layer found, try a different approach
    if last_conv_layer is None:
        # Look for layers with 'conv' in name
        for layer in reversed(model.layers):
            if hasattr(layer, 'name') and 'conv' in layer.name.lower():
                last_conv_layer = layer
                break
        
        # If still not found, use a specific layer name that should exist
        if last_conv_layer is None:
            try:
                # Try common layer names from InceptionV3/EfficientNet
                possible_names = ['mixed10', 'mixed9', 'top_conv', 'block7a_expand_conv']
                for name in possible_names:
                    try:
                        last_conv_layer = model.get_layer(name)
                        break
                    except:
                        continue
            except:
                pass
    
    if last_conv_layer is None:
        print("No suitable convolutional layer found for Grad-CAM")
        return None
    
    print(f"Using layer: {last_conv_layer.name}")
    
    # Create gradient model
    try:
        grad_model = tf.keras.models.Model(
            model.inputs, [last_conv_layer.output, model.output]
        )
    except Exception as e:
        print(f"Error creating gradient model: {e}")
        return None
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if class_index is None:
            class_index = tf.argmax(predictions[0])
        loss = predictions[:, class_index]
    
    # Compute gradients
    grads = tape.gradient(loss, conv_outputs)
    if grads is None:
        print("Could not compute gradients")
        return None
        
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Generate heatmap
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0)
    
    # Normalize heatmap
    max_val = tf.reduce_max(heatmap)
    if max_val > 0:
        heatmap = heatmap / max_val
    
    return heatmap.numpy()

# Cell 18: Visualize Grad-CAM
def visualize_gradcam(image_path):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return
    
    # Load and preprocess image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    # Get prediction
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class]
    
    # Generate Grad-CAM
    heatmap = get_gradcam(model, img_array, predicted_class)
    
    if heatmap is not None:
        # Resize heatmap
        heatmap = cv2.resize(heatmap, (img.size[0], img.size[1]))
        heatmap = np.uint8(255 * heatmap)
        heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
        
        # Overlay on original image
        original_img = np.array(img)
        superimposed_img = cv2.addWeighted(original_img, 0.6, heatmap, 0.4, 0)
        
        # Plot results
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.imshow(original_img)
        plt.title("Original Image")
        plt.axis('off')
        
        plt.subplot(1, 3, 2)
        plt.imshow(heatmap)
        plt.title("Grad-CAM Heatmap")
        plt.axis('off')
        
        plt.subplot(1, 3, 3)
        plt.imshow(superimposed_img)
        plt.title(f"Overlay\nPredicted: Class {predicted_class}\nConfidence: {confidence:.2f}")
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()

# Cell 19: Alternative simpler Grad-CAM function
def simple_gradcam(model, img_array, layer_name=None):
    """
    Simplified Grad-CAM implementation that works better with complex models
    """
    # If no layer specified, try to find a suitable one
    if layer_name is None:
        # Try common layer names
        possible_layers = ['mixed10', 'mixed9', 'mixed8', 'top_conv', 'block7a_expand_conv', 'block6a_expand_conv']
        for name in possible_layers:
            try:
                test_layer = model.get_layer(name)
                layer_name = name
                print(f"Found layer: {layer_name}")
                break
            except:
                continue
    
    if layer_name is None:
        print("Could not find suitable layer for Grad-CAM")
        return None
    
    # Create a model that maps the input image to the activations of the last conv layer
    grad_model = tf.keras.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )
    
    # Compute the gradient of the top predicted class for our input image
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    
    # Gradient of the output neuron (top predicted) with regard to the output feature map
    grads = tape.gradient(class_channel, last_conv_layer_output)
    
    # Mean intensity of the gradient over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Multiply each channel in the feature map array by "how important this channel is"
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    # Normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# Cell 20: Updated visualization function
def visualize_gradcam_simple(image_path, layer_name=None):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return
    
    # Load and preprocess image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    # Get prediction
    prediction = model.predict(img_array, verbose=0)
    predicted_class = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class]
    
    # Get class names if available
    class_names = list(train_generator.class_indices.keys()) if 'train_generator' in globals() else [f"Class {i}" for i in range(len(prediction[0]))]
    
    print(f"Predicted: {class_names[predicted_class]} (Confidence: {confidence:.3f})")
    
    # Generate Grad-CAM
    heatmap = simple_gradcam(model, img_array, layer_name)
    
    if heatmap is not None:
        # Resize heatmap to match original image size
        heatmap_resized = cv2.resize(heatmap, (img.size[0], img.size[1]))
        heatmap_colored = np.uint8(255 * heatmap_resized)
        heatmap_colored = cv2.applyColorMap(heatmap_colored, cv2.COLORMAP_JET)
        
        # Convert original image to array
        original_img = np.array(img)
        
        # Superimpose the heatmap on original image
        superimposed_img = heatmap_colored * 0.4 + original_img * 0.6
        superimposed_img = np.clip(superimposed_img, 0, 255).astype(np.uint8)
        
        # Display results
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.imshow(original_img)
        plt.title("Original Image")
        plt.axis('off')
        
        plt.subplot(1, 3, 2)
        plt.imshow(heatmap, cmap='jet')
        plt.title("Grad-CAM Heatmap")
        plt.axis('off')
        
        plt.subplot(1, 3, 3)
        plt.imshow(superimposed_img)
        plt.title(f"Overlay\n{class_names[predicted_class]}\nConf: {confidence:.3f}")
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
    else:
        print("Could not generate Grad-CAM visualization")

# Example usage - try both functions
sample_image_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/normal/normal (1).png'

print("Trying original Grad-CAM function:")
try:
    visualize_gradcam(sample_image_path)
except Exception as e:
    print(f"Error with original function: {e}")

print("\nTrying simplified Grad-CAM function:")
try:
    visualize_gradcam_simple(sample_image_path)
except Exception as e:
    print(f"Error with simplified function: {e}")
```

```python
# Cell 1: Install packages and mount drive
!pip install tensorflow keras numpy opencv-python matplotlib scikit-learn flask seaborn

from IPython import get_ipython
from IPython.display import display
import os

# Cell 2: Import libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3, EfficientNetB0
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Concatenate, GlobalAveragePooling2D, Input
from tensorflow.keras.models import Model
from tensorflow.keras.utils import Sequence
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score
import seaborn as sns

# Cell 3: List directories to debug path
project_path = "/content/drive/MyDrive/cancer project"
if os.path.exists(project_path):
    print("\nDirectories in cancer project:")
    for item in os.listdir(project_path):
        print(item)
else:
    print("cancer project does not exist")

# Cell 4: Verify dataset path
dataset_path = "/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT"
if os.path.exists(dataset_path):
    print(f"\ndataset path is correct and contains: {os.listdir(dataset_path)}")
else:
    print("dataset path does not exist, update the path")

# Cell 5: Image parameters
IMG_SIZE = 224
BATCH_SIZE = 16
EPOCHS = 10

# Cell 6: Image Augmentation - Fixed validation split issue
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.6, 1.4],
    fill_mode="nearest"
)

val_datagen = ImageDataGenerator(rescale=1.0 / 255.0)

# Cell 7: Load training & validation data - Fixed data loading
# Assuming you have separate train/val folders or create them
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    dataset_path,  # You might need separate validation folder
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

print(f"Number of classes: {train_generator.num_classes}")
print(f"Class indices: {train_generator.class_indices}")

# Cell 8: Create ensemble model with proper input handling
def create_ensemble_model(num_classes):
    # Single input layer
    input_layer = Input(shape=(224, 224, 3))
    
    # Load pretrained models
    inception_base = InceptionV3(weights='imagenet', include_top=False, input_tensor=input_layer)
    efficientnet_base = EfficientNetB0(weights='imagenet', include_top=False, input_tensor=input_layer)
    
    # Freeze pretrained layers initially
    inception_base.trainable = False
    efficientnet_base.trainable = False
    
    # Extract features
    inception_features = GlobalAveragePooling2D()(inception_base.output)
    efficientnet_features = GlobalAveragePooling2D()(efficientnet_base.output)
    
    # Merge features
    merged = Concatenate()([inception_features, efficientnet_features])
    merged = BatchNormalization()(merged)
    merged = Dense(512, activation='relu')(merged)
    merged = Dropout(0.3)(merged)
    merged = Dense(256, activation='relu')(merged)
    merged = Dropout(0.2)(merged)
    output = Dense(num_classes, activation='softmax')(merged)
    
    model = Model(inputs=input_layer, outputs=output)
    return model

# Create the model
model = create_ensemble_model(train_generator.num_classes)

# Cell 9: Compile model with correct loss function
model.compile(
    optimizer=Adam(learning_rate=0.0001), 
    loss='categorical_crossentropy',  # Fixed: was binary_crossentropy
    metrics=['accuracy']
)

model.summary()

# Cell 10: Train the model (initial training)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=[early_stopping]
)

# Cell 11: Fine-tuning - Unfreeze some layers
# Unfreeze last 20 layers of both models
for layer in model.layers:
    if hasattr(layer, 'layers'):  # Check if it's a nested model
        for sublayer in layer.layers[-20:]:
            sublayer.trainable = True

# Compile with lower learning rate
model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Cell 12: Fine-tuning training
fine_tune_epochs = 10
callbacks = [
    EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
]

history_fine = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=fine_tune_epochs,
    callbacks=callbacks
)

# Cell 13: Plot training results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
if 'history' in locals():
    plt.plot(history.history['accuracy'], label='Initial Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Initial Val Accuracy')
plt.plot(history_fine.history['accuracy'], label='Fine-tune Train Accuracy')
plt.plot(history_fine.history['val_accuracy'], label='Fine-tune Val Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
if 'history' in locals():
    plt.plot(history.history['loss'], label='Initial Train Loss')
    plt.plot(history.history['val_loss'], label='Initial Val Loss')
plt.plot(history_fine.history['loss'], label='Fine-tune Train Loss')
plt.plot(history_fine.history['val_loss'], label='Fine-tune Val Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()

# Cell 14: Create test set for evaluation
# Get all image paths and labels
image_paths = train_generator.filepaths
labels = train_generator.labels

# Convert categorical labels to one-hot if needed
from tensorflow.keras.utils import to_categorical
labels_categorical = to_categorical(labels, num_classes=train_generator.num_classes)

# Load and preprocess images
X = []
for image_path in image_paths:
    img = cv2.imread(image_path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    X.append(img)

X = np.array(X, dtype='float32') / 255.0

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, labels_categorical, test_size=0.2, random_state=42, stratify=labels
)

# Cell 15: Comprehensive Model Evaluation
def evaluate_model_comprehensive(model, X_test, y_test, class_names):
    """
    Comprehensive evaluation including accuracy, confusion matrix, F1 score, precision, recall
    """
    print("=" * 60)
    print("COMPREHENSIVE MODEL EVALUATION")
    print("=" * 60)
    
    # Get predictions
    y_pred_proba = model.predict(X_test, verbose=0)
    y_pred = np.argmax(y_pred_proba, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    # Basic metrics
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"\nBasic Metrics:")
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Test Loss: {test_loss:.4f}")
    
    # Detailed classification report
    print(f"\nDetailed Classification Report:")
    print("-" * 50)
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    print(classification_report(y_true, y_pred, target_names=class_names))
    
    # Extract metrics for each class
    print(f"\nPer-Class Metrics:")
    print("-" * 50)
    for i, class_name in enumerate(class_names):
        precision = report[class_name]['precision']
        recall = report[class_name]['recall']
        f1 = report[class_name]['f1-score']
        support = report[class_name]['support']
        
        print(f"{class_name}:")
        print(f"  Precision: {precision:.4f}")
        print(f"  Recall:    {recall:.4f}")
        print(f"  F1-Score:  {f1:.4f}")
        print(f"  Support:   {int(support)}")
        print()
    
    # Overall metrics
    print(f"Overall Metrics:")
    print("-" * 20)
    macro_precision = precision_score(y_true, y_pred, average='macro')
    macro_recall = recall_score(y_true, y_pred, average='macro')
    macro_f1 = f1_score(y_true, y_pred, average='macro')
    
    weighted_precision = precision_score(y_true, y_pred, average='weighted')
    weighted_recall = recall_score(y_true, y_pred, average='weighted')
    weighted_f1 = f1_score(y_true, y_pred, average='weighted')
    
    print(f"Macro Average:")
    print(f"  Precision: {macro_precision:.4f}")
    print(f"  Recall:    {macro_recall:.4f}")
    print(f"  F1-Score:  {macro_f1:.4f}")
    print()
    print(f"Weighted Average:")
    print(f"  Precision: {weighted_precision:.4f}")
    print(f"  Recall:    {weighted_recall:.4f}")
    print(f"  F1-Score:  {weighted_f1:.4f}")
    
    return y_true, y_pred, y_pred_proba

def plot_confusion_matrix(y_true, y_pred, class_names, normalize=True):
    """
    Plot confusion matrix with customization options
    """
    # Calculate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # Create subplots for both normalized and raw confusion matrices
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Raw confusion matrix
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names, ax=axes[0])
    axes[0].set_title('Confusion Matrix (Raw Counts)')
    axes[0].set_xlabel('Predicted Label')
    axes[0].set_ylabel('True Label')
    
    # Normalized confusion matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_normalized, annot=True, fmt='.3f', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names, ax=axes[1])
    axes[1].set_title('Normalized Confusion Matrix')
    axes[1].set_xlabel('Predicted Label')
    axes[1].set_ylabel('True Label')
    
    plt.tight_layout()
    plt.show()
    
    # Print confusion matrix statistics
    print("\nConfusion Matrix Analysis:")
    print("-" * 30)
    total_samples = np.sum(cm)
    correct_predictions = np.trace(cm)
    
    for i, class_name in enumerate(class_names):
        true_positives = cm[i, i]
        false_positives = np.sum(cm[:, i]) - true_positives
        false_negatives = np.sum(cm[i, :]) - true_positives
        true_negatives = total_samples - true_positives - false_positives - false_negatives
        
        print(f"\n{class_name}:")
        print(f"  True Positives:  {true_positives}")
        print(f"  False Positives: {false_positives}")
        print(f"  False Negatives: {false_negatives}")
        print(f"  True Negatives:  {true_negatives}")
    
    return cm

def plot_class_distribution(y_true, y_pred, class_names):
    """
    Plot class distribution comparison
    """
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # True distribution
    unique_true, counts_true = np.unique(y_true, return_counts=True)
    axes[0].bar([class_names[i] for i in unique_true], counts_true, color='skyblue', alpha=0.7)
    axes[0].set_title('True Class Distribution')
    axes[0].set_ylabel('Count')
    axes[0].tick_params(axis='x', rotation=45)
    
    # Predicted distribution
    unique_pred, counts_pred = np.unique(y_pred, return_counts=True)
    axes[1].bar([class_names[i] for i in unique_pred], counts_pred, color='lightcoral', alpha=0.7)
    axes[1].set_title('Predicted Class Distribution')
    axes[1].set_ylabel('Count')
    axes[1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

# Get class names
class_names = list(train_generator.class_indices.keys())

# Perform comprehensive evaluation
print("Starting comprehensive evaluation...")
y_true, y_pred, y_pred_proba = evaluate_model_comprehensive(model, X_test, y_test, class_names)

# Plot confusion matrix
print("\nGenerating confusion matrix...")
cm = plot_confusion_matrix(y_true, y_pred, class_names)

# Plot class distributions
print("\nGenerating class distribution plots...")
plot_class_distribution(y_true, y_pred, class_names)

# Cell 16: Additional Evaluation Metrics
def calculate_additional_metrics(y_true, y_pred_proba, class_names):
    """
    Calculate additional metrics like AUC-ROC for multi-class
    """
    from sklearn.metrics import roc_auc_score, roc_curve, auc
    from sklearn.preprocessing import label_binarize
    
    print("\nAdditional Metrics:")
    print("-" * 20)
    
    # Convert to binary format for multi-class ROC
    y_true_binary = label_binarize(y_true, classes=range(len(class_names)))
    
    # Calculate AUC for each class
    if len(class_names) > 2:
        try:
            # Multi-class AUC
            auc_scores = []
            for i in range(len(class_names)):
                if len(np.unique(y_true_binary[:, i])) > 1:  # Check if class exists in test set
                    auc_score = roc_auc_score(y_true_binary[:, i], y_pred_proba[:, i])
                    auc_scores.append(auc_score)
                    print(f"AUC-ROC for {class_names[i]}: {auc_score:.4f}")
                else:
                    print(f"AUC-ROC for {class_names[i]}: N/A (class not in test set)")
            
            if auc_scores:
                macro_auc = np.mean(auc_scores)
                print(f"Macro Average AUC-ROC: {macro_auc:.4f}")
        except Exception as e:
            print(f"Could not calculate AUC-ROC: {e}")
    else:
        # Binary classification AUC
        try:
            auc_score = roc_auc_score(y_true, y_pred_proba[:, 1])
            print(f"AUC-ROC: {auc_score:.4f}")
        except Exception as e:
            print(f"Could not calculate AUC-ROC: {e}")

# Calculate additional metrics
calculate_additional_metrics(y_true, y_pred_proba, class_names)

# Cell 17: Save detailed results
def save_evaluation_results(y_true, y_pred, y_pred_proba, class_names, model_name="breast_cancer_model"):
    """
    Save detailed evaluation results to files
    """
    import pandas as pd
    from datetime import datetime
    
    # Create results dictionary
    results = {
        'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        'model_name': model_name,
        'total_samples': len(y_true),
        'accuracy': np.mean(y_true == y_pred),
        'class_names': class_names
    }
    
    # Calculate per-class metrics
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    
    # Create DataFrame for detailed results
    detailed_results = []
    for class_name in class_names:
        if class_name in report:
            detailed_results.append({
                'class': class_name,
                'precision': report[class_name]['precision'],
                'recall': report[class_name]['recall'],
                'f1_score': report[class_name]['f1-score'],
                'support': report[class_name]['support']
            })
    
    results_df = pd.DataFrame(detailed_results)
    print("\nDetailed Results Summary:")
    print(results_df.to_string(index=False))
    
    # Save confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    cm_df = pd.DataFrame(cm, index=class_names, columns=class_names)
    
    print(f"\nConfusion Matrix:")
    print(cm_df.to_string())
    
    return results_df, cm_df

# Save results
results_df, cm_df = save_evaluation_results(y_true, y_pred, y_pred_proba, class_names)

# Cell 18: Save model
model.save('breast_cancer_ensemble_model.keras')
print("Model saved successfully!")

# Cell 19: Grad-CAM visualization function (keeping original)
def get_gradcam(model, img_array, class_index=None):
    # Get the last convolutional layer by checking layer names and types
    last_conv_layer = None
    
    # First, try to find conv layers in the base models
    for layer in model.layers:
        if hasattr(layer, 'layers'):  # This is a nested model (InceptionV3 or EfficientNet)
            for sublayer in reversed(layer.layers):
                if 'conv' in sublayer.name.lower() and hasattr(sublayer, 'output'):
                    try:
                        output_shape = sublayer.output.shape
                        if len(output_shape) == 4:  # Check if it's a conv layer
                            last_conv_layer = sublayer
                            break
                    except:
                        continue
            if last_conv_layer:
                break
    
    # If no conv layer found, try a different approach
    if last_conv_layer is None:
        # Look for layers with 'conv' in name
        for layer in reversed(model.layers):
            if hasattr(layer, 'name') and 'conv' in layer.name.lower():
                last_conv_layer = layer
                break
        
        # If still not found, use a specific layer name that should exist
        if last_conv_layer is None:
            try:
                # Try common layer names from InceptionV3/EfficientNet
                possible_names = ['mixed10', 'mixed9', 'top_conv', 'block7a_expand_conv']
                for name in possible_names:
                    try:
                        last_conv_layer = model.get_layer(name)
                        break
                    except:
                        continue
            except:
                pass
    
    if last_conv_layer is None:
        print("No suitable convolutional layer found for Grad-CAM")
        print("Available layers:")
        for i, layer in enumerate(model.layers):
            print(f"{i}: {layer.name} - {type(layer).__name__}")
        return None
    
    print(f"Using layer: {last_conv_layer.name}")
    
    # Create gradient model
    try:
        grad_model = tf.keras.models.Model(
            model.inputs, [last_conv_layer.output, model.output]
        )
    except Exception as e:
        print(f"Error creating gradient model: {e}")
        return None
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if class_index is None:
            class_index = tf.argmax(predictions[0])
        loss = predictions[:, class_index]
    
    # Compute gradients
    grads = tape.gradient(loss, conv_outputs)
    if grads is None:
        print("Could not compute gradients")
        return None
        
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Generate heatmap
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0)
    
    # Normalize heatmap
    max_val = tf.reduce_max(heatmap)
    if max_val > 0:
        heatmap = heatmap / max_val
    
    return heatmap.numpy()

# Cell 20: Enhanced visualization with metrics
def visualize_gradcam_with_metrics(image_path, y_true_idx=None):
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return
    
    # Load and preprocess image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    
    # Get prediction
    prediction = model.predict(img_array, verbose=0)
    predicted_class = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class]
    
    # Get class names if available
    class_names = list(train_generator.class_indices.keys()) if 'train_generator' in globals() else [f"Class {i}" for i in range(len(prediction[0]))]
    
    print(f"Predicted: {class_names[predicted_class]} (Confidence: {confidence:.3f})")
    if y_true_idx is not None:
        print(f"True Label: {class_names[y_true_idx]}")
        print(f"Correct Prediction: {'Yes' if predicted_class == y_true_idx else 'No'}")
    
    # Show all class probabilities
    print("\nAll Class Probabilities:")
    for i, prob in enumerate(prediction[0]):
        print(f"  {class_names[i]}: {prob:.3f}")
    
    # Generate simple Grad-CAM (using the working function)
    heatmap = simple_gradcam(model, img_array)
    
    if heatmap is not None:
        # Resize heatmap to match original image size
        heatmap_resized = cv2.resize(heatmap, (img.size[0], img.size[1]))
        heatmap_colored = np.uint8(255 * heatmap_resized)
        heatmap_colored = cv2.applyColorMap(heatmap_colored, cv2.COLORMAP_JET)
        
        # Convert original image to array
        original_img = np.array(img)
        
        # Superimpose the heatmap on original image
        superimposed_img = heatmap_colored * 0.4 + original_img * 0.6
        superimposed_img = np.clip(superimposed_img, 0, 255).astype(np.uint8)
        
        # Display results
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.imshow(original_img)
        plt.title("Original Image")
        plt.axis('off')
        
        plt.subplot(1, 3, 2)
        plt.imshow(heatmap, cmap='jet')
        plt.title("Grad-CAM Heatmap")
        plt.axis('off')
        
        plt.subplot(1, 3, 3)
        plt.imshow(superimposed_img)
        title = f"Overlay\n{class_names[predicted_class]}\nConf: {confidence:.3f}"
        if y_true_idx is not None:
            title += f"\n{'' if predicted_class == y_true_idx else ''}"
        plt.title(title)
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
    else:
        print("Could not generate Grad-CAM visualization")

# Cell 21: Alternative simpler Grad-CAM function (keeping original)
def simple_gradcam(model, img_array, layer_name=None):
    """
    Simplified Grad-CAM implementation that works better with complex models
    """
    # If no layer specified, try to find a suitable one
    if layer_name is None:
        # Try common layer names
        possible_layers = ['mixed10', 'mixed9', 'mixed8', 'top_conv', 'block7a_expand_conv', 'block6a_expand_conv']
        for name in possible_layers:
            try:
                test_layer = model.get_layer(name)
                layer_name = name
                print(f"Found layer: {layer_name}")
                break
            except:
                continue
    
    if layer_name is None:
        print("Could not find suitable layer. Available layers:")
        for layer in model.layers:
            print(f"- {layer.name}")
        return None
    
    # Create a model that maps the input image to the activations of the last conv layer
    grad_model = tf.keras.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )
    
    # Compute the gradient of the top predicted class for our input image
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    
    # Gradient of the output neuron (top predicted) with regard to the output feature map
    grads = tape.gradient(class_channel, last_conv_layer_output)
    
    # Mean intensity of the gradient over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Multiply each channel in the feature map array by "how important this channel is"
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    
    # Normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

# Example usage with enhanced metrics
sample_image_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/normal/normal (1).png'

print("Enhanced Grad-CAM visualization with metrics:")
try:
    visualize_gradcam_with_metrics(sample_image_path)
except Exception as e:
    print(f"Error with enhanced visualization: {e}")

print("\n" + "="*60)
print("EVALUATION COMPLETE")
print("="*60)
print(f"Final Test Accuracy: {test_accuracy:.4f}")
print(f"Macro F1-Score: {f1_score(y_true, y_pred, average='macro'):.4f}")
print(f"Weighted F1-Score: {f1_score(y_true, y_pred, average='weighted'):.4f}")
print("Model and evaluation results saved successfully!")
```
